```python
#important command to display IMMEDIATELY your plots
%matplotlib inline
#-*- coding : utf-8-*-
# coding:unicode_escape

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

**Global COVID-19 dataset**

```python
url='https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
df = pd.read_csv(url,index_col=0)
df.head()
```

### $\color{#00C5CD}{SouthAmerica}$<br>**New covid cases in South America from 2020-02-22 to 2022-11-18**




```python
df_SouthAmerica = df[df['location'] == 'South America']
df_SouthAmerica[['date', 'new_cases']].plot(kind='line',
                                       x='date',
                                       y='new_cases',
                                       title='new cases in South America',
                                       legend=False)
plt.show()
```

This chart shows the number of new COVID-19 cases diagnosed per million people in South America between February 22, 2020 and November 18, 2022. Counts can include probable cases, where reported. It can be seen from the table that the overall trend of change in South America is unstable and has a recurring trend. It surged and peaked between January and March 2022 at about 490,000 per million people; The number of new cases decreased sharply between March and May 2022 and reached a minimum around December 2021, 2 May 2022 and November 2022, with about 100,000 new cases per million people.

**Total cases per million in South America from 2020-02-22 to 2022-11-18**

```python
#total_cases per million
df_SouthAmerica3 = df[df['location'] == 'South America']
df_SouthAmerica3[['date', 'total_cases_per_million']].plot(kind='line',
                                       x='date',
                                       y='total_cases_per_million',
                                       title='total cases per million in South America',
                                       legend=False)
plt.show()
```

**Total deaths per million in South America from 2020-02-22 to 2022-11-18**

This table shows the total number of confirmed COVID-19 cases per 1 million people in South America between 22 February 2020 and 18 November 2022. As can be seen from the table, the overall change of the number of diagnosed cases in South America shows a gentle upward trend, which is more obvious between July 2021 and January 2021 and August 2022 and November 2022. However, between December 2021 and March 2022, the number of diagnosed cases showed a sharp rise.

```python
df_SouthAmerica4 = df[df['location'] == 'South America']
df_SouthAmerica4[['date', 'total_deaths_per_million']].plot(kind='line',
                                       x='date',
                                       y='total_deaths_per_million',
                                       title='total deaths per million in South America',
                                       legend=False)
plt.show()
```

This table shows the total number of COVID-19 deaths per one million people in South America between February 22, 2020 and November 18, 2022. It can be seen from the table that the total number of deaths is relatively small, and the overall trend is rising. Before July 2021, there will be a rapid upward trend, and after July 2021, there will be a gradual upward trend, that is, fewer and fewer people will die from COVID-19. From another perspective, it also reflects that the relevant policies and medical services in South American countries are becoming more and more in place, gradually controlling the death rate of covid-19.

### **Next, the graphs below are about total vaccinations per hundred in South America and North America**

```python
#only focus on South America
df_S = df[df['location'] == 'South America']
df_S[['date', 'total_vaccinations_per_hundred']].plot(kind='line',
                                       x='date',
                                       y='total_vaccinations_per_hundred',
                                       title='total vaccinations per hundred in South America',
                                       legend=False)

plt.xticks(rotation=60)
plt.show()
```

```python
#only focus on North America
df_North = df[df['location'] == 'North America']
df_North[['date', 'total_vaccinations_per_hundred']].plot(kind='line',  #kind: by default is line
                                       x='date',
                                       y='total_vaccinations_per_hundred',
                                       title='total vaccinations per hundred in North America',
                                       legend=False)

plt.xticks(rotation=60)
plt.show()
```

The above two tables respectively show the total number of COVID-19 vaccinations per 100 people in the total population of South America and North America. Through comparative analysis of the two tables, it can be seen that the total number of COVID-19 vaccinations in South America and North America is on the rise as a whole. In addition, the number of people vaccinated showed a sharp increase between April 2021 and April 2022. By comparing the death trend charts of South America and North America mentioned above, it can be seen that the death trend of South America due to COVID-19 became relatively gentle after April 2022. <font color=blue size=3>This means that vaccination has played a role in inhibiting COVID-19 in South America.</font>

### **Total vaccinations per hundred in five continents: South America,North America,Asia, Africa, and Europe**

```python
df_5locations = df.loc[df['location'].isin(['South America', 'North America','Asia','Africa','Europe'])]
df_5locations_pivot = df_5locations.pivot_table(values='total_vaccinations_per_hundred', columns='location', index='date') #each location with one column
df_5locations_pivot.plot(title='total_vaccinations_per_hundred in different continents')
plt.xticks(rotation=60)
plt.show()
```

This table shows the total number of vaccinations per 100 people across the five continents (South America, North America, Asia, Africa and Europe) between December 2, 2020 and November 2, 2022. Through comparative analysis, <font color=blue size=3>it can be concluded that the vaccination situation of the five continents is on the rise.</font> <br><br>
However, the trend of vaccination gradually showed polarization after January 2021, with the other four continents except Africa showing a sharp upward trend between January 2021 and February 2022, and vaccination was more active. Africa has seen a flattening of vaccination trends, with the lowest total number of vaccinations per 100 people in the five continents and the highest total number of vaccinations per 100 people at just 50.<br><br> The situation of the other four continents was as follows: At the beginning, the vaccination rate in Asia was the lowest and that in South America was the second from the bottom. However, after October 2021, the vaccination rate in South America surpassed the other continents, and finally reached the second place in Asia and the continent with the highest vaccination rate in the five continents.<font color=blue size=3>This shows that the vaccination response in South America and Asia is better.</font>




### **Compare life expectancy, GDP, human development index on different continents**<br>
- Human Development Index<br>
A composite index measuring average achievement in three basic dimensions of human development—a long and healthy life, knowledge and a decent standard of living.

```python
for i in ('life_expectancy','gdp_per_capita','human_development_index'):
    df_all_pivot_l = df.pivot_table(values=i, columns='continent', index='date')
    df_all_pivot_l.plot(title=i + ' in Different Continents') # line plot
    plt.legend(loc='lower right')

plt.show()
```

In general, all continents were affected, and there has been a sharp drop in life expectancy, GDP, and HDI after Covid-19. <br>
Overall, life expectancy, GDP and human development index are the highest in Europe and the lowest in Africa, with the other continents in the middle.
Combined with the previous chart, vaccination rates in North and South America and Asia are on the rise.<font color=blue size=3> It can be indirectly concluded that vaccination rates will affect life expectancy, human development index and reduce the impact of covid-19 on people.</font>


### **Comparing GDP per capita of different continents**

```python
df_pivot = df.pivot_table(values='gdp_per_capita', columns='continent', index='location')
df_pivot.plot(kind='hist', alpha=0.3, bins=10, title='GDP Per Capita by Continent')
plt.show()
```

We can see that African and Asian countries have the largest proportion of per capita GDP below 20,000, European countries have the largest proportion of per capita GDP between 20,000 and 50,000, and a few European and Asian countries have per capita GDP above 80,000

### **Comparing stringency_index of different continents**<br>
- Government Response Stringency Index<br>
Composite measure based on 9 response indicators including school closures, workplace closures, and travel bans, rescaled to a value from 0 to 100 (100 = strictest response)

```python
df_stringency_pivot = df.pivot_table(values='stringency_index', columns='continent', index='location')
sns.boxplot(data=df_stringency_pivot, palette='rainbow').set_title('Stringency by Continent')
plt.show()
```

This box plot indicates the distribution of the stringency index in different continents. We can intuitively see that the median of South America is higher than other continents. And the stringency of Europe,North America and South America have outliers. Besides, in the case of the most stringent policy response, the index was highest in Asia.
